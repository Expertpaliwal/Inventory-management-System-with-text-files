```python
import time

fd=open("Inventory.txt","r")

txt=fd.read()

fd.close()
products=txt.split("\n")

print("Product List are as follows    :")
print("-----------------------------------------")
for i in products:
    print(i)
print("-----------------------------------------")


ui_username = input("Enter your Name:  ")
ui_phone=input("Enter your phone number :")
ui_prod_id=input("Enter product Id: ")
ui_prod_qn=input("Enter product Quantity: ")

updated_product=[]

for product in products:
    productDetails=product.split(",")
    if ui_prod_id== productDetails[0]:
        if(int(ui_prod_qn)<=int(productDetails[3])):
            print("--------------------------------------------------")
            print("Product Name             :",productDetails[1])
            print("Price                    :",productDetails[2])
            print("Quantity                 :",ui_prod_qn)
            print("--------------------------------------------------")
            print("Billing Amount           :",int(productDetails[2])*int(ui_prod_qn))
            print("--------------------------------------------------")

            productDetails[3]=str(int(productDetails[3])-int(ui_prod_qn))
            
            
            fd= open("sales.txt","a")
            user_detail=ui_username+","+ui_phone+","+productDetails[1]+","+ui_prod_qn+","+time.ctime()+","+str(int(productDetails[2])*int(ui_prod_qn))+"\n"
            fd.write(user_detail)
            fd.close()
        else:
            print("Sorry, We're not having Enough Quantity!!!!")
            print("Present Quantity of the Item : ", productDetails[3])
            print("Would you like to purchase it: ")
            
            ch=input("Press Y/N:  ")
            if(ch=='Y' or ch=='y'):
                print("--------------------------------------------------")
                print("Product Name             :",productDetails[1])
                print("Price                    :",productDetails[2])
                print("Quantity                 :",productDetails[3])
                print("--------------------------------------------------")
                print("Billing Amount           :",int(productDetails[2])*int(productDetails[3]))
                print("--------------------------------------------------")
                productDetails[3]='0'
            else:
                print("Thanks")
                        
    updated_product.append(productDetails)

fd=open("inventory.txt","w")
i=0
for j in updated_product:
    prod=j[0]+","+j[1]+","+j[2]+","+j[3]
    if(i!=len(updated_product)-1):
        prod+="\n"
    i+=1
    fd.write(prod)
fd.close()

print("-----------------------------------\n Inventory Updated")
```

```python
def calculateBill(prod_id,qn,productDetails):
    print("--------------------------------------------------")
    print("Product Name             :",productDetails[1])
    print("Price                    :",productDetails[2])
    print("Quantity                 :",qn)
    print("--------------------------------------------------")
    print("Billing Amount           :",int(productDetails[2])*qn)
    print("--------------------------------------------------")
    return str(int(productDetails[2])*qn)
```

```python
print("----------------------------------------------")
print("Welcome to the Inventory Management")
print("----------------------------------------------")
import time

fd=open("Inventory.txt","r")

txt=fd.read()

fd.close()

products=txt.split("\n")

print("Product List are as follows    :")
print("-----------------------------------------")
for i in products:
    print(i)
print("-----------------------------------------")


print("Enter Your Details !!! ")
ui_username = input("Enter your Name:  ")
ui_phone=input("Enter your phone number :")
print("Enter What you want to purchase!!! ")
ui_prod_id=[]
ui_prod_qn=[]
while(True):
    uid=int(input("Enter product Id: "))
    qn=int(input(f"Enter product Quantity for {uid}: "))
    ui_prod_id.append(uid)
    ui_prod_qn.append(qn)
    
    ch= input("Do you want to purchase more items: (y/n)")
    if(ch=='y' or ch=='Y'):
        continue
    else:
        break

updated_product=[]        

if(len(ui_prod_id)!=0):
    for product in products:
        
        productDetails=product.split(",")
        
        if int(productDetails[0]) in ui_prod_id:
            
            ind=ui_prod_id.index(int(productDetails[0]))
            
            if(ui_prod_qn[ind]<=int(productDetails[3])):
                
                amount= calculateBill(ui_prod_id[ind],ui_prod_qn[ind],productDetails)
                
                productDetails[3]=str(int(productDetails[3])-ui_prod_qn[ind])
                
                
                fd= open("sales.txt","a")
                
                user_detail=ui_username+","+ui_phone+","+productDetails[1]+","+str(ui_prod_qn[ind])+","+time.ctime()+","+amount+"\n"
                fd.write(user_detail)
                fd.close()
            
            else:
                
                print("Sorry, We're not having Enough Quantity!!!!")
                print("Present Quantity of the Item : ", productDetails[3])
                print("Would you like to purchase it: ")

                ch=input("Press Y/N:  ")
                if(ch=='Y' or ch=='y'):
                    amount= calculateBill(ui_prod_id[ind],productDetails[3],productDetails)
                    productDetails[3]=0
                    
                    fd= open("sales.txt","a")
                    user_detail=ui_username+","+ui_phone+","+productDetails[1]+","+str(ui_prod_qn[ind])+","+time.ctime()+","+amount+"\n"
                    fd.write(user_detail)
                    fd.close()
                
                else:
                    print("Thanks")
                    
        updated_product.append(productDetails)
    fd=open("inventory.txt","w")
    i=0
    for j in updated_product:
        prod=j[0]+","+j[1]+","+j[2]+","+j[3]
        if(i!=len(updated_product)-1):
            prod+="\n"
        i+=1
        fd.write(prod)
    fd.close()
    print("-----------------------------------\n Inventory Updated")
        
            
print("--------------------------------------------------\nThanks\n-----------------------------------------------")
```
